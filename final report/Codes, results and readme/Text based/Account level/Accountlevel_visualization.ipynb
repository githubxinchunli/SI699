```python
import pandas as pd
users_lr=pd.read_csv('users_lr.csv')
users_rf=pd.read_csv('users_rf.csv')
users_svm=pd.read_csv('users_svm.csv')
users_boosting=pd.read_csv('users_boosting.csv')
random=pd.read_csv('random.csv')
logistic=pd.read_csv('logistic.csv')
boosting=pd.read_csv('boost.csv')
svm=pd.read_csv('result_svm.csv')
```

```python
import numpy as np


users=pd.DataFrame(np.array([[users_svm[users_svm['add']==0]['micro'].median(),
                         users_svm[users_svm['add']==100]['micro'].median(),
                         users_svm[users_svm['add']==200]['micro'].median(),
                         users_svm[users_svm['add']==300]['micro'].median(),
                         users_svm[users_svm['add']==400]['micro'].median()],
                         [users_rf[users_rf['add']==0]['micro'].median(),
                         users_rf[users_rf['add']==100]['micro'].median(),
                         users_rf[users_rf['add']==200]['micro'].median(),
                         users_rf[users_rf['add']==300]['micro'].median(),
                         users_rf[users_rf['add']==400]['micro'].median()],
                         [users_boosting[users_boosting['add']==0]['micro'].median(),
                          users_boosting[users_boosting['add']==100]['micro'].median(),
                          users_boosting[users_boosting['add']==200]['micro'].median(),
                          users_boosting[users_boosting['add']==300]['micro'].median(),
                          users_boosting[users_boosting['add']==400]['micro'].median()],
                         [users_lr[users_lr['add']==0]['micro'].median(),
                          users_lr[users_lr['add']==100]['micro'].median(),
                          users_lr[users_lr['add']==200]['micro'].median(),
                          users_lr[users_lr['add']==300]['micro'].median(),
                          users_lr[users_lr['add']==400]['micro'].median()]]),
                columns=['No Add', 'Add 100', 'Add 200', 'Add 300', 'Add 400'],
               index = ['SVM', 'RF', 'GDBoosting', 'LR'])
```

```python

import matplotlib.pyplot as plt

plt.plot(users.iloc[0], color='cyan', label='SVM')
plt.plot(users.iloc[1], 'blue', label='Random Forest')
plt.plot(users.iloc[2], 'yellow', label='Gradient Boosting')
plt.plot(users.iloc[3], 'red', label='Logistic Regression')
plt.title('Median accuracy of adding portions of test data to train set ')  


plt.legend()  
plt.ylabel('Accuracy')
plt.rcParams['figure.figsize'] = (7.0, 5.0) 

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.subplot(1,4,1)
ax = sns.distplot(users_svm['micro'], rug=True, hist=True)
ax.set_title('SVM accuracy histogram')
ax.set_xlabel('Accuracy')
ax.set_ylabel('Frequency')

plt.subplot(1,4,2)
ax = sns.distplot(users_rf['micro'], rug=True, hist=True)
ax.set_title('Random Forest accuracy histogram')
ax.set_xlabel('Accuracy')
ax.set_ylabel('Frequency')

plt.subplot(1,4,3)
ax = sns.distplot(users_boosting['micro'], rug=True, hist=True)
ax.set_title('Gradient Boosting accuracy histogram')
ax.set_xlabel('Accuracy')
ax.set_ylabel('Frequency')

plt.subplot(1,4,4)
ax = sns.distplot(users_lr['micro'], rug=True, hist=True)
ax.set_title('Logistic Regression accuracy histogram')
ax.set_xlabel('Accuracy')
ax.set_ylabel('Frequency')

plt.subplots_adjust(left=0.1 , bottom=0.1, right=2.2, top=0.8,
                wspace=0.2, hspace=0.35)
```
